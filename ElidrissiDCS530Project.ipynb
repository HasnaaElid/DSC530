```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("student_data.csv")
data.head()
```

### <font color = "green">Description of Feilds chosen for the analysis </font>

| Field Name |   Description                                     |   Type  |        Possible values                                       |
|------------|---------------------------------------------------|---------|---------------------------------------------------------------------------|
| sex        | student’s sex                                     |binary| ‘F’ - female or ‘M’ - male                                           |
| age        |student’s age                                      |numeric| from 15 to 22                                                     |
| famsize | family size |binary| ‘LE3’ - less or equal to 3 or ‘GT3’ - greater than 3|
| Pstatus | parent’s cohabitation status |binary| ‘T’ - living together or ‘A’ - apart|
|Medu |mother’s education |numeric| 0 - none, 1 - primary education (4th grade), 2 - 5th to 9th grade, 3 - secondary education or 4 - higher education)|
|fedu |mother’s education |numeric| 0 - none, 1 - primary education (4th grade), 2 - 5th to 9th grade, 3 - secondary education or 4 - higher education)|
|Fjob |father’s job |nominal| ‘teacher’, ‘health’ care related, civil ‘services’ (e.g. administrative or police), ‘at_home’ or ‘other’|
|guardian |student’s guardian |nominal| ‘mother’, ‘father’ or ‘other’|
|studytime |weekly study time |numeric| 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)|
|failures |number of past class failures |numeric| n if 1<=n<3, else 4|
|schoolsup |extra educational support |binary| yes or no|
|famsup |family educational support |binary| yes or no
|G3 | final grade |numeric| from 0 to 20, output target|

```python
data = data[["sex","age","famsize","Pstatus","Medu","Fedu","Fjob","guardian","studytime","failures","schoolsup","famsup","G3"]]
data.head(10)
```

### <font color= "green">Statistical Question</font>  

- Does parental education level (Medu, Fedu) afect student performance (G3 final grade)?

### <font color="green">Variables & Hypothesis</font>

```python
data.describe()
```

#### <font color="green">Dependent Varialble</font>

- sex	student’s sex
- age	student’s age
- famsize	family size
- Pstatus	parent’s cohabitation status
- Medu	mother’s education
- Fjob	father’s job
- guardian	student’s guardian
- studytime	weekly study time
- failures	number of past class failures
- schoolsup	extra edcational support
- famsup	family educational support

#### <font color= "green">Independent variables<font>

- G3	final grade

#### <font color="green">Hypothesis</font>

- Null Hypothesis (H0): Parental education level, study time, and number of past failures do not significantly affect the student’s final grade (G3).
- Alternative Hypothesis (H1): Parental education level, study time, and fewer past failures positively affect the student’s final grade (G3).

## <font color="green">Descriptive Analysis of Variables</font>

```python
data.isna().sum()
```

```python
data.describe()
```

## <font color = "green">Distributions</font>

```python
gray = "#7D7D7D"
green = "#7D9D7D"
for column in ['age','Medu', 'Fedu','studytime','failures', 'G3']:
    sns.histplot(data[column], kde=True,color=gray)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

## <font color= "green">PMF Comparison</font>

```python
pmf_high = data[(data['Medu'] >= 3) & (data['Fedu'] >= 3)]['G3'].value_counts(normalize=True)
pmf_low = data[(data['Medu'] < 3) & (data['Fedu'] < 3)]['G3'].value_counts(normalize=True)

pmf_high.sort_index().plot(kind='bar', color=green, alpha=0.7, label='High Parental Education')
pmf_low.sort_index().plot(kind='bar', color=gray, alpha=0.7, label='Low Parental Education')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Probability')
plt.title('PMF of Final Grade by Parental Education Level')
plt.legend()
plt.show()
```

## <font color="green">CDF (Cumulative Distribution Function) of G3</font>

```python
sorted_data = data['G3'].sort_values()
cdf = sorted_data.value_counts(normalize=True).sort_index().cumsum()
plt.plot(cdf.index, cdf.values,color = gray, alpha=1)
plt.xlabel('Final Grade (G3)')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Final Grade (G3)')
plt.show()
```

## <font color="green"> Analytical distribution</font>

```python
mean_g3 = data['G3'].mean()
std_g3 = data['G3'].std()

plt.hist(data['G3'], bins=10, density=True, alpha=0.6, color='g', edgecolor='black')

x = np.linspace(data['G3'].min(), data['G3'].max(), 100)
pdf = stats.norm.pdf(x, mean_g3, std_g3)

plt.plot(x, pdf, 'k', color=gray, linewidth=2, label="Normal Distribution Fit")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Density")
plt.title("Fitting a Normal Distribution to Final Grade (G3)")
plt.legend()
plt.show()
```

## <font color="green">Scatter Plot Analysis</font>

```python
sns.scatterplot(x='Medu', y='G3', data=data, color= green)
plt.title('Mother’s Education Level vs. Final Grade (G3)')
plt.show()

sns.scatterplot(x='Fedu', y='G3', data=data, color = green)
plt.title('Father’s Education Level vs. Final Grade (G3)')
plt.show()
```

## <font color="green">Hypothesis Testing</font>

```python
high_education_g3 = data[(data['Medu'] >= 3) & (data['Fedu'] >= 3)]['G3']
low_education_g3 = data[(data['Medu'] < 3) & (data['Fedu'] < 3)]['G3']
t_stat, p_value = stats.ttest_ind(high_education_g3, low_education_g3, equal_var=False)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

## <font color="green">Regression Analysis</font>

```python
X = data[['Medu', 'Fedu']]
y = data['G3']

# Add a constant to the model
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()
print(model.summary())
```
